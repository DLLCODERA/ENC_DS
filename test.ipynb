```python
import tensorflow as tf
```

```python
a = tf.random.normal([200])
print(a)
a = tf.expand_dims(a, -1)
print(a)
```

```python
a = tf.keras.layers.Input(128)
a = tf.expand_dims(a, -1)
print(a)
```

```python
beta = tf.random.normal([200])
beta = tf.expand_dims(beta, -1)
print(beta)
```

```python
gama = tf.keras.layers.Input(11)
print(gama)
# 不包尾 gama[:, 0:-1] = gama[:, 0:11]
print(gama[:,0:-1])
# 包尾
print(gama[:,0:10])
```
